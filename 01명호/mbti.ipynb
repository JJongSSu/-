```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random

from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname = 'C:\Windows\Fonts\malgun.ttf').get_name()
rc('font',family=font_name)

import warnings
warnings.filterwarnings('ignore')
```

```python
red = pd.read_csv('사업 기초 세우기 - 설문지 응답 시트1.csv',sep=',',encoding='utf-8')
red
```

```python
red.info()
```

```python
plt.figure(figsize=(12,12))
plt.pie(red.iloc[:,2].value_counts(),autopct="%1.2f%%", 
        colors = ['#8fd9b6', '#ff9999','#ffc000', '#d395d0', '#FAFAD2', '#FFE4C4','#E0FFFF','#FFF0F5'], 
        explode = [0.01,0.01,0.01,0.01,0.01,0.01,0.01], counterclock=False
        ,startangle=280,radius=0.8)
plt.legend(red.iloc[:,2].value_counts().keys(),loc='upper right')
title_font = {
    'fontsize': 15,
    'fontweight': 'bold'
}
plt.title('유기농/비건/친환경 제품의 정보는 어떻게 구하시나요?', pad=10, fontdict=title_font)
plt.savefig('qu2.png')
```

```python
## 데이터 준비
labels = red.iloc[:,2].value_counts().keys() ## 라벨
frequency = red.iloc[:,2].value_counts() ## 빈도
 
fig = plt.figure(figsize=(10,10)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 배경색을 하얀색으로 설정
ax = fig.add_subplot() ## 프레임 생성
 
pie = ax.pie(frequency, ## 파이차트 출력
       startangle=90, ## 시작점을 90도(degree)로 지정
       counterclock=False, ## 시계 방향으로 그린다.
       autopct=lambda p : '{:.2f}%'.format(p) ## 퍼센티지 출력
       )

plt.legend(pie[0],labels,loc='upper right') ## 범례 표시
plt.show()
```

```python
red.iloc[:,2].value_counts().plot(figsize=(10,10), kind='pie', autopct='%.2f%%', 
                                  colors = ['#8fd9b6', '#ff9999','#ffc000', '#d395d0', '#FAFAD2', '#FFE4C4','#E0FFFF','#FFF0F5'], 
                                  explode = [0.01,0.01,0.01,0.01,0.01,0.01,0.01], counterclock=False, fontsize=12, radius=0.8)
plt.legend(loc='upper right', fontsize=8)
plt.savefig('qu1.png')
```
