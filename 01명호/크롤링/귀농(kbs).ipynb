```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
from tqdm import tqdm_notebook as tqdm
from bs4 import BeautifulSoup as bs
import requests as req
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
driver.get('https://news.kbs.co.kr/search/search.do?query=%EA%B7%80%EB%86%8D#1')
```

```python
url_list = []
for i in range(1,100):
    page = 'https://news.kbs.co.kr/search/search.do?query=%EA%B7%80%EB%86%8D#{}'.format(i)
    driver.get(page)
    time.sleep(1)

    soup1 = bs(driver.page_source,'html.parser')
    search = soup1.select('a')
    for i in search:
        url_list.append(i['href'])
len(url_list)
```

```python
# 상세 페이지 기사 제목 & 내용
title_list = []
contents_list = []

for url in tqdm(url_list):
    driver.get(url)
    time.sleep(1)
    soup = bs(driver.page_source,'html.parser')  
    title = soup.select('div.w_article_title > h3')
    contents = soup.select('div.text_area')
    for i in title:
        title_list.append(i.text)
    # title_list.append(title[0].text)
    for i in contents:
        contents_list.append(i.text)  
    # contents_list.append(contents[0].text)
len(title_list), len(contents_list)
```

```python
newsdata = {'제목':title_list,'참조주소':url_list,'내용':contents_list}
current_news = pd.DataFrame(newsdata)
current_news
```
