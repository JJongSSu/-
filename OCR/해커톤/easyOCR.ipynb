```python
from pdf2image.pdf2image import convert_from_path
import easyocr

pdf_path = './test.pdf'
images = convert_from_path(pdf_path, fmt='jpeg')

reader = easyocr.Reader(['ko','en'])
extracted_text = []

for image in images:
    result = reader.readtext(image)
    extracted_text.extend([text for text, _, _ in result])
    
print(extracted_text)

# extracted_text 리스트 내의 리스트 요소를 문자열로 변환
cleaned_text = []

for item in extracted_text:
    if isinstance(item, list):
        text = item[0]  # 리스트의 첫 번째 요소를 추출
    else:
        text = item  # 이미 문자열인 경우 그대로 사용
    cleaned_text.append(text)

print("cleaned_text", cleaned_text)

# 변환된 텍스트를 파일에 쓰기
with open('output.txt', 'w', encoding='utf-8') as output_file:
    output_file.write('\n'.join(cleaned_text))
```
