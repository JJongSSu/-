```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
# import matplotlib.pylab -> 시각화 + numpy
```

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y, ls = '--', marker = 'o', lw = 3, c = 'black', mfc = 'r', mec = 'r', mew = 4, ms = 5)   
# c(color) : 선 색깔
# ls 종류 :  '--', '-.', ':'
# marker 종류 : 'o','<','>','s', '*', 'D'
# lw : 선 굵기
# mfc(markerfacecolor) : 점 내부 색깔(테두리는 안변함)
# ms : marker size
# mec : 마커 선 색깔
# mew : 마커 선 굵기
plt.show()
```

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y, c = 'g', marker = 'o', ls = '--', lw = 3, mew = 5, ms = 7)
plt.show
```

```python
plt.plot(x,y, c = 'm', marker = 'D', ls = ':', lw = 2, mew = 2, ms = 7, mfc = 'b', mec = 'r')
```

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y, c='b', marker = '<', ls='-.', lw = 2, ms = 15, mfc = 'g', mew = 2, mec = 'r')

# x,y 축 범위 설정
plt.xlim(-3,10)
plt.ylim(-2,10)

# 격자무늬
plt.grid()  

plt.show()
```

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y, c='b', marker = '<', ls='-.', lw = 2, ms = 15, mfc = 'g', mew = 2, mec = 'r')

# x,y 축 범위 설정
plt.xticks(range(0,10,2))
plt.yticks([-2,4,10])

# 격자무늬
plt.grid()  

plt.show()
```

```python
# 기본적으로 y축 기준으로 그래프 그려짐
x = [1,2,3,4]
y = [2,4,6,8]
z = [1,3,5,7]
plt.plot(x, label = '엑스')
plt.plot(y, label = '와이')
plt.plot(z, label = '제트')
# 범례 -> plot에 label 넣어줘야 됨
plt.legend()    
plt.show()
```
