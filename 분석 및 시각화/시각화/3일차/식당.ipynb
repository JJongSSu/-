```python
a = df[['quantity','item_price']]
new_price_data = []
for i in range(len(a['item_price'])):
    new_price_data.append(a['item_price'][i].replace('$',''))
new_price = pd.Series(new_price_data)
new_price = new_price.astype("float64")
a['new_price'] = new_price
a
```
