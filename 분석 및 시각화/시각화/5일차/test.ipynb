```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname = 'C:\Windows\Fonts\malgun.ttf').get_name()
rc('font',family=font_name)
```

```python
x = [2,3,4,5,6,7,8]
y1 = [5,7,3,9,1,8,6]
y2 = [9,5,7,2,3,7,1]

plt.plot(x,y1, label='Y1', c='r', lw = 2)
plt.plot(x,y2, label='Y2', c='b', lw = 2)

plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('<실행결과>')
plt.legend()    
plt.show()
```

```python
tf = pd.read_csv('Traffic_Accident_2017.csv',sep=',', encoding='cp949')
tf.head()
```

```python
tf['주야'] = tf['주야'].str.replace('야간','Night')
tf['주야'] = tf['주야'].str.replace('주간','Day')
tf['주야']
```

```python
LAT = tf['위도']
LON = tf['경도']
tf.head()
plt.figure(figsize=(8,5))
sns.scatterplot(x=LON,y=LAT,hue='주야',data=tf, alpha=0.1, palette='Set1')
plt.xlim(123,133)
plt.xlabel('LON')
plt.ylabel('LAT')
plt.show()
```
