```python
# 한솥도시락 페이지에서 메뉴 이름, 가격 수집
driver = webdriver.Chrome('C:\Study\데이터 수집, 웹 개요\chromedriver.exe')
driver.get('https://www.hsd.co.kr/')
```

```python
driver.maximize_window()
```

```python
menu = driver.find_element_by_xpath('//*[@id="gnb"]/div[2]/ul/li[3]/p/a')
```

```python
from selenium.webdriver.common.action_chains import ActionChains
```

```python
# action chain 만들기
actions = ActionChains(driver)
actions.move_to_element(menu)
all_menu = driver.find_element_by_xpath('//*[@id="gnb"]/div[2]/ul/li[3]/div/ul/li[1]/a')
actions.click(all_menu)
actions.perform()
```

```python
# 더보기 버튼 누르기 : 더보기 모두 누른 후에 data를 가져와야 됨(아니면 data가 계속 중복되서 가져와짐)
while True:
    try:
        plus_btn = driver.find_element_by_css_selector('#btn_more > span > a')
        plus_btn.click()
    except:
        break
```

```python
name = driver.find_elements_by_css_selector('h4')
for i in name:
    print(i.text)
```

```python
price = driver.find_elements_by_css_selector('div.item-price > strong')
for i in price:
    print(i.text)
```

```python
driver = webdriver.Chrome('C:\Study\데이터 수집, 웹 개요\chromedriver.exe')
driver.get('https://www.hsd.co.kr/')
driver.maximize_window()
```

```python
menu = driver.find_element_by_xpath('//*[@id="gnb"]/div[2]/ul/li[3]/p/a')
all_menu = driver.find_element_by_xpath('//*[@id="gnb"]/div[2]/ul/li[3]/div/ul/li[1]/a')
```

```python
actions = ActionChains(driver)
actions.move_to_element(menu)
actions.click(all_menu)
actions.perform()
```

```python
soup = bs(driver.page_source, 'html.parser')
```

```python
name_bs = soup.select('h4.h.fz_03')
price_bs = soup.select('div.item-price > strong')
```

```python
name_bs_list = []
price_bs_list = []
for i in name_bs:
    name_bs_list.append(i.text)

for i in price_bs:
    price_bs_list.append(i.text)
```
