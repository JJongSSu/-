```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
from tqdm import tqdm
# from tqdm import tqdm_notebook as tqdm
from bs4 import BeautifulSoup as bs
import requests as req
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import csv
import random
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
driver.get('https://www.paiksool.com/')
```

```python
driver.maximize_window()
```

```python
driver.find_element('xpath','//*[@id="wrap"]/header/div[1]/ul/li[1]').click()
time.sleep(1)

driver.find_element('xpath','//*[@id="contents"]/div/div/div[2]/div/ul/li[2]').click()
time.sleep(1)
```

```python
total_list = []
title_list = []
second_title_list = []
evaluation_list=[]
content_list=[]

for i in tqdm(range(1,29)):
    page = 'https://www.paiksool.com/goodsinfo/goods_list.php?page={}&cateCd=019'.format(i)
    driver.get(page)
    time.sleep(2)

    soup = bs(driver.page_source, 'html.parser')
    total = soup.select('#container > div.goods_list_item > div.list_prd > ul')
    
    total_list.append(total)

    for i in tqdm(range(1,11)):
        # Image = driver.find_element(By.CSS_SELECTOR,'#container > div.goods_list_item > div.list_prd > ul > li:nth-child({}) > a > div.box_img > img'.format(i))
        # Image.click()
        driver.find_element(By.CSS_SELECTOR,'#container > div.goods_list_item > div.list_prd > ul > li:nth-child({}) > a > div.box_img > img'.format(i)).click()
        
        # title = total[i].select('#container > div.section.s01 > div > div.box_txt_drink > h3')[0].text
        title = driver.find_element(By.CSS_SELECTOR,'#container > div.section.s01 > div > div.box_txt_drink > h3').text
        title_list.append(title)
        time.sleep(2)

        second_title = driver.find_element(By.CSS_SELECTOR,'#container > div.section.s01 > div > div.box_txt_drink > p').text
        second_title_list.append(second_title)
        time.sleep(2)
        
        try:
            for i in range(1,4):
                try:
                    evaluation = driver.find_element(By.CSS_SELECTOR,f'#container > div.section.s03 > div > ul > li:nth-child({i}) > p')
                    evaluation_list.append(evaluation.text)
                except:
                    evaluation = ''
                    evaluation_list.append(evaluation)
        except:
            evaluation = ''
            evaluation_list.append(evaluation)
            time.sleep(2)
        
        try:
            content = driver.find_element(By.CSS_SELECTOR,'#container > div.section.s04 > div > ul.con > div.tab_con1')
            content_list.append(content.text)
        except:
            content = ''
            content_list.append(content)
            time.sleep(2)           
        driver.get(page)
        time.sleep(2)
    
len(total_list), len(title_list), len(second_title_list), len(evaluation_list), len(content_list)
```

```python
data = {'이름':title_list,'설명':second_title_list,'평가':evaluation_list_new,'내용':content_list}
paiksool = pd.DataFrame(data)
paiksool
```
