```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
from tqdm import tqdm
# from tqdm import tqdm_notebook as tqdm
from bs4 import BeautifulSoup as bs
import requests as req
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
driver.get('https://www.paiksool.com/')
```

```python
driver.maximize_window()
```

```python
driver.find_element('xpath','//*[@id="wrap"]/header/div[1]/ul/li[1]').click()
time.sleep(1)

driver.find_element('xpath','//*[@id="contents"]/div/div/div[2]/div/ul/li[2]').click()
time.sleep(1)
```

```python
total_list = []
title_list = []
second_title_list = []
evaluation_list=[]
content_list=[]

for i in tqdm(range(1)):
    page = 'https://www.paiksool.com/goodsinfo/goods_list.php?page={}&cateCd=019'.format(i)
    driver.get(page)
    time.sleep(2)

    soup = bs(driver.page_source, 'html.parser')
    total = soup.select('#container > div.goods_list_item > div.list_prd > ul')

    for i in tqdm(range(len(total))):
        title = total[i].select('#container > div.goods_list_item > div.list_prd > ul > li:nth-child(1) > a > div.txt_box > p.name.ellipsis')[0].text
        title_list.append(title)
        time.sleep(3)

        second_title = total[i].select('#container > div.goods_list_item > div.list_prd > ul > li:nth-child(1) > a > div.txt_box > p.exple_box')[0].text
        second_title_list.append(second_title)
        time.sleep(2)

    total_list.append(total)

    for i in tqdm(range(1,len(total)+1)):
        Image = driver.find_element_by_xpath('//*[@id="container"]/div[3]/div[2]/ul/li[{}]/a'.format(i))
        Image.click()

        try:
            evaluation = driver.find_element_by_css_selector('#container > div.section.s03 > div > ul > li:nth-child(1) > p')
            evaluation_list.append(evaluation.text)
        except:
            evaluation = ''
            evaluation_list.append(evaluation)
            time.sleep(2)
        
        try:
            content = driver.find_element_by_css_selector('#container > div.section.s04 > div > ul.con > div.tab_con1')
            content_list.append(content.text)
        except:
            content = ''
            content_list.append(content)
            time.sleep(3)           
        driver.get(page)
        time.sleep(2)
    
len(total_list), len(title_list), len(second_title_list), len(evaluation_list), len(content_list)
```

```python
data = {'이름':title_list,'설명':second_title_list,'평가':evaluation_list,'내용':content_list}
paiksool = pd.DataFrame(data)
paiksool
```
