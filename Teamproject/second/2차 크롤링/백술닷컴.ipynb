```python
stars_list = []
heart_list = []

for i in tqdm(range(24)):
    page = 'https://www.musinsa.com/search/musinsa/goods?q=%ED%96%A5%EC%88%98&list_kind=small&sortCode=pop&sub_sort=&page={}&display_cnt=0&saleGoods=&includeSoldOut=&setupGoods=&popular=&category1DepthCode=&category2DepthCodes=&category3DepthCodes=&selectedFilters=&category1DepthName=&category2DepthName=&brandIds=&price=&colorCodes=&contentType=&styleTypes=&includeKeywords=&excludeKeywords=&originalYn=N&tags=&campaignId=&serviceType=&eventType=&type=&season=&measure=&openFilterLayout=N&selectedOrderMeasure=&shoeSizeOption=&groupSale=&d_cat_cd=&attribute='.format(i)
    driver.get(page)
    time.sleep(2)

    for i in tqdm(range(len(total))):
        Image = driver.find_elements_by_css_selector('img.lazyload.lazy')
        Image[i].click()
        try:
            stars = driver.find_element_by_css_selector('a.prd-score__link > span.prd-score__rating')
            stars_list.append(stars.text)
        except:
            stars = '0'
            stars_list.append(stars)

        try:
            heart = driver.find_element_by_css_selector('span.prd_like_cnt')
            heart_list.append(heart.text)
        except:    
            heart = '0'
            heart_list.append(heart)
            time.sleep(2)           
        driver.get(page)
        time.sleep(2)

len(stars_list), len(heart_list)
```
