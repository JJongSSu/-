```python
import pandas as pd
best = pd.read_csv('Best50.csv')
best = pd.DataFrame(best)
forhim = pd.read_csv('ForHim.csv')
formen = pd.DataFrame(forhim)
forher = pd.read_csv('ForHer.csv')
forwomen = pd.DataFrame(forher)
unisex = pd.read_csv('Unisex.csv')
unisex = pd.DataFrame(unisex)
```

```python
perfume = pd.concat([best,unisex,formen,forwomen],keys = ['Best','Unisex','Men','Women'])
perfume.columns = ['No','이름','브랜드','향 설명','가격','평점']
perfume.index.name = '성별'
perfume = perfume.drop(['No'],axis = 1)
price= perfume['가격'].str.split(' ').str[0]
perfume['가격'] = price
perfume.to_csv('perfume.csv')
```
