```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import time
from tqdm import tqdm_notebook as tqdm

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re

from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname = 'C:\Windows\Fonts\malgun.ttf').get_name()
rc('font',family=font_name)

import warnings
warnings.filterwarnings('ignore')
```

```python
perfume = pd.read_csv('perfume.csv',encoding='utf-8')
musinsa = pd.read_csv('musinsa_woman.csv',encoding='utf-8')
```

```python
perfume.columns = ['구분','삭제','이름','브랜드','설명','가격','평점']
perfume = perfume.drop(['삭제'],axis=1)
perfume
```

```python
musinsa
```

```python
musinsa.columns = ['No','브랜드','이름','가격','평점','하트']
musinsa.index = musinsa.No
musinsa = musinsa.drop(['No'],axis=1)
musinsa
```

```python
price = []
price1 = []
for i in range(len(musinsa['가격'])):
    imsi = musinsa['가격'][i].replace('\n','')
    imsi = imsi.strip()
    price.append(imsi)
for i in range(len(price)):
    price1.append(price[i][-7:])
price1
```

```python
price[5][-7:]
```

```python
musinsa['가격'] = price1
musinsa
```

```python
musinsa['이름'][0]
```

```python
name = []
for i in range(len(musinsa['이름'])):
    imsi1 = musinsa['이름'][i].replace('\n','')
    while "  " in imsi1:
        imsi1 = imsi1.replace("  ","")
        name.append(imsi1)
musinsa['이름'] = name
musinsa
```

```python
musinsa['구분'] = 'Women'
musinsa = musinsa[['구분','이름','브랜드','가격','평점','하트']]
musinsa
```

```python
total = pd.concat([perfume,musinsa], join='outer', axis=0, ignore_index=True)
total
```

```python
total.to_csv('향수사이트.csv')
```
