```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("향수검색.csv", encoding = "utf-8-sig")
df
```

```python
# time series decomposition
# 2017년 9월 27일은 수요일 (요일 = 0이면 수요일)

day_of_week = []
for i in range(365):
    for w in range(6):
        day_of_week.append(i)
df["요일"] = day_of_week[:1832]
df
```

```python
# trend
trend = []
for a, i in enumerate(df["향수"]):
    try:
        trend.append(np.mean(df["향수"].loc[a-182 : a+182]))
    except:
        trend.append(np.nan)
trend
```

```python
for i in range(182):
    trend[i] = np.nan
    trend[(i+1)*(-1)] = np.nan
trend
```

```python
detrend = []
for a, i in enumerate(trend):
    detrend.append(df["향수"].loc[a] - i)
detrend
```

```python
plt.plot(trend)
```

```python
# season
df["디트렌드"] = detrend
season = []
for i in range(6):
    imsi = df[df["연도"] == i]
    season.append(np.mean(imsi["디트렌드"]))
    
df["주기"] = df["연도"].apply(lambda x : season[x])
df
```

```python
season
```

```python
plt.plot(range(5), season[:5])
```

```python
df["잔차"] = df["디트렌드"] - df["주기"]
plt.plot(df["잔차"])
```
